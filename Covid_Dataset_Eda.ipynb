# Data Visualization (For Exploratory Data Analysis)

# Top 5 countries with confirmed covid cases

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract top 5 countries by confirmed cases
top_5_countries = df.sort_values(by='Confirmed', ascending=False).head(5)

# Create a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Confirmed', y='Country/Region', data=top_5_countries, palette='viridis')
plt.xlabel('Confirmed Cases')
plt.ylabel('Country/Region')
plt.title('Top 5 Countries by Total Confirmed COVID-19 Cases')
plt.tight_layout()

# Display number of cases on the y-axis
for i, (country, cases) in enumerate(zip(top_5_countries['Country/Region'], top_5_countries['Confirmed'])):
    plt.text(cases, i, f' {cases}', ha='left', va='center', fontsize=10)

plt.show()
```

# Top 5 countries with recovered covid cases

```python
# Extract top 5 countries by recovered cases
top_5_recovered = df.sort_values(by='Recovered', ascending=False).head(5)

# Create a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Recovered', y='Country/Region', data=top_5_recovered, palette='viridis')
plt.xlabel('Recovered Cases')
plt.ylabel('Country/Region')
plt.title('Top 5 Countries by Total Recovered COVID-19 Cases')
plt.tight_layout()

# Display number of cases on the y-axis
for i, (country, recovered) in enumerate(zip(top_5_recovered['Country/Region'], top_5_recovered['Recovered'])):
    plt.text(recovered, i, f' {recovered}', ha='left', va='center', fontsize=10)

plt.show()
```

# Top 5 countries with most deaths by covid

```python
# Extract top 5 countries by deaths
top_5_deaths = df.sort_values(by='Deaths', ascending=False).head(5)

# Create a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Deaths', y='Country/Region', data=top_5_deaths, palette='viridis')
plt.xlabel('Deaths')
plt.ylabel('Country/Region')
plt.title('Top 5 Countries by Total COVID-19 Deaths')
plt.tight_layout()

# Display number of deaths on the y-axis
for i, (country, deaths) in enumerate(zip(top_5_deaths['Country/Region'], top_5_deaths['Deaths'])):
    plt.text(deaths, i, f' {deaths}', ha='left', va='center', fontsize=10)

plt.show()
```

# Top 10 countries with highest recovery rates

```python
# Calculate recovery rates
df['Recovery Rate (%)'] = (df['Recovered'] / df['Confirmed']) * 100

# Filter countries with significant data (e.g., more than 1000 confirmed cases)
significant_data = df[df['Confirmed'] > 1000].sort_values(by='Recovery Rate (%)', ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Recovery Rate (%)', y='Country/Region', data=significant_data, palette='muted')
plt.xlabel('Recovery Rate (%)')
plt.ylabel('Country/Region')
plt.title('Top 10 Countries by Recovery Rate (%)')
plt.tight_layout()

# Display numbers on the bars
for i, (country, rate) in enumerate(zip(significant_data['Country/Region'], significant_data['Recovery Rate (%)'])):
    plt.text(rate, i, f' {rate:.2f}%', ha='left', va='center', fontsize=10)

plt.show()
```

# Top 10 countries with highest mortality rate

```python
# Calculate mortality rates
df['Mortality Rate (%)'] = (df['Deaths'] / df['Confirmed']) * 100

# Filter countries with significant data (e.g., more than 1000 confirmed cases)
significant_data = df[df['Confirmed'] > 1000].sort_values(by='Mortality Rate (%)', ascending=False).head(10)

# Create a bar plothttp://localhost:8888/notebooks/Covid_Dataset_Eda.ipynb#
plt.figure(figsize=(12, 8))
sns.barplot(x='Mortality Rate (%)', y='Country/Region', data=significant_data, palette='rocket')
plt.xlabel('Mortality Rate (%)')
plt.ylabel('Country/Region')
plt.title('Top 10 Countries by Mortality Rate (%)')
plt.tight_layout()

# Display numbers on the bars
for i, (country, rate) in enumerate(zip(significant_data['Country/Region'], significant_data['Mortality Rate (%)'])):
    plt.text(rate, i, f' {rate:.2f}%', ha='left', va='center', fontsize=10)

plt.show()
```

# Distribution of confirmed cases by WHO region

```python
# Calculate total confirmed cases by WHO region
region_data = df.groupby('WHO Region').sum()

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(region_data['Confirmed'], labels=region_data.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3'))
plt.title('Distribution of Confirmed COVID-19 Cases by WHO Region')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
